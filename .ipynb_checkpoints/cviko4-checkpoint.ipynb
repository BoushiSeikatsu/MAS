```python
import pandas as pd
import seaborn as sns
import networkx as nx
import numpy as np
import re
import matplotlib.pyplot as plt
import itertools
```

```python
def createGraphFromCSV(path, sep):
    listOfConnections = []
    with open(path, "r") as file:
        lines = file.readlines()
    for line in lines:
        regexString = "(\d*)"+sep+"(\d*)"
        match = re.match(regexString, line)
        if match is not None: 
            connection = (match.group(1),match.group(2))
            listOfConnections.append(connection)
        else:
            continue
    return listOfConnections
```

```python
listOfConnections = createGraphFromCSV("E:\\Python Scripts\\MAS\\KarateClub.csv", sep=';')
```

```python
def createSubGraph(vertices, originalConnections):
    listOfConnections = []
    for vertex in vertices:
        for connection in originalConnections:
            if(connection[0] == vertex):
                if(connection[1] in vertices):
                    listOfConnections.append((connection[0], connection[1]))
            elif(connection[1] == vertex):
                if(connection[0] in vertices):
                    listOfConnections.append((connection[1], connection[0]))
    return listOfConnections
```

```python
def drawCommunities(G, communities):
    community_map = {}
    for i, community in enumerate(communities):
        for node in community:
            community_map[node] = i

    # Create a color list where the color of each node is based on its community
    colors = [community_map[node] for node in G.nodes()]

    # Visualize the graph with nodes colored by community
    plt.figure(figsize=(10, 8))
    pos = nx.spring_layout(G, seed=42)  # positions for all nodes

    # Draw the nodes with community-based colors
    nx.draw_networkx_nodes(G, pos, node_color=colors, cmap=plt.cm.rainbow, node_size=300)

    # Draw the edges
    nx.draw_networkx_edges(G, pos, alpha=0.5)

    # Add labels (optional)
    nx.draw_networkx_labels(G, pos)

    plt.title('Graph with Communities')
    plt.show()
```

```python
def charactarizeCommunities(communities):
    totalSet = {}
    i = 0
    for set in communities:
        totalSet[i] = list(set)
        i += 1
    processedList = []
    #print(totalSet)
    checkVal = 0 # check value, in case the vertex is not in graph, their value will be NaN
    for i in range(1,len(df) + 1):
        checkVal = 0
        for key, values in totalSet.items():
            if str(i) in values:
                processedList.append(key)
                checkVal = 1
                break
        if checkVal == 0:
            processedList.append('NaN')
    return processedList

def charactarizeCommunitiesClique(communities):
    totalSet = {}
    i = 0
    for set in communities:
        totalSet[i] = list(set)
        i += 1
    processedList = []
    for i in range(1,len(df) + 1):
        processedList.append([])
        for key, values in totalSet.items():
            if str(i) in values:
                processedList[i - 1].append(key)
    return processedList
```

```python
df['Louivan'] = charactarizeCommunities(louivanCommunity)
```

```python
df['Label Propagation'] = charactarizeCommunities(LPCommunity)
```

```python
df['Kerninghan'] = charactarizeCommunities(KLBCommunity)
```

```python
df['Girwan Newman'] = charactarizeCommunities(GNCommunity)
```

```python
df['K=4 Clique'] = charactarizeCommunitiesClique(KCliqueCommunity)
```
